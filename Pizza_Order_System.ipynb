```python
#Pizza Üst Sınıfı Tanımla
class Pizza:
    def __init__(self):
        self.description = ""
        self.cost = 0.0
    def get_description(self):
        return self.description
    def get_cost(self):
        return self.cost
```

```python
#Pizza Alt Sınıflarını Tanımla
class KlasikPizza(Pizza):
    def __init__(self):
        super().__init__()
        self.description = "Klasik Pizza(Biber,Domates)"
        self.cost = 10.0

class MargaritaPizza(Pizza):
    def __init__(self):
        super().__init__()
        self.description = "Margarita pizza(Mozerella,Biber,Domates)"
        self.cost = 12.0

class TurkPizza(Pizza):
    def __init__(self):
        super().__init__()
        self.description = "Turk pizzasi(Sucuk,Kasar,Acı Biber,Domates)"
        self.cost = 15.0

class SadePizza(Pizza):
    def __init__(self):
        super().__init__()
        self.description = "Sade pizza"
        self.cost = 8.0
```

```python
#Decorator Üst Sınıfı Tanımla
class Decorator(Pizza):
    def __init__(self, component):
        super().__init__()
        self.component = component
    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self)
    def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self)

#Decorator Sos Alt Sınıflarını Tanımla
class Zeytin(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self.description = "Zeytinli"
        self.cost = 2.0
    def get_cost(self):
        return self.component.get_cost() + self.cost
    def get_description(self):
        return self.component.get_description() + ' ' + self.description

class Mantarlar(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self.description = "Mantarlarlı"
        self.cost = 3.0
    def get_cost(self):
        return self.component.get_cost() + self.cost
    def get_description(self):
        return self.component.get_description() + ' ' + self.description

class KeciPeyniri(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self.description = "Keçi Peynirili"
        self.cost = 4.0
    def get_cost(self):
        return self.component.get_cost() + self.cost
    def get_description(self):
        return self.component.get_description() + ' ' + self.description

class Et(Decorator):
    def __init__(self,component):
     super().__init__(component)
     self.description = "Etli"
     self.cost = 10.0
    def get_cost(self):
        return self.component.get_cost() + self.cost
    def get_description(self):
        return self.component.get_description() + ' ' + self.description

class Sogan(Decorator):
    def __init__(self,component):
     super().__init__(component)
     self.description = "Soganlı"
     self.cost = 2.5
    def get_cost(self):
        return self.component.get_cost() + self.cost
    def get_description(self):
        return self.component.get_description() + ' ' + self.description

class Misir(Decorator):
    def __init__(self,component):
     super().__init__(component)
     self.description = "Mısırlı"
     self.cost = 1.5
    def get_cost(self):
        return self.component.get_cost() + self.cost
    def get_description(self):
        return self.component.get_description() + ' ' + self.description
```

```python
# datetime ve csv Kütüphanelerini Import Et
import datetime
import csv


#Main Method'unu tanımla 
def main():
    # Menu.txt dosyasını oku ve menüyü ekrana yazdır
    with open("Menu.txt", "r") as f:
        menu = f.read()
        print(menu)

    # Kullanıcıdan pizza seçimini al
    while True:
        try:
            pizza_choice = int(input("Lütfen bir pizza seçin (1-4): "))
            if 1 <= pizza_choice <= 4:
                break
            else:
                print("Geçersiz seçim. Lütfen 1-4 arasında bir sayı girin.")
        except ValueError:
            print("Geçersiz seçim. Lütfen 1-4 arasında bir sayı girin.")

    # Kullanıcıdan sos seçimini al
    while True:
        try:
            sos_choice = int(input("Lütfen bir sos seçin (11-16): "))
            if 11 <= sos_choice <= 16:
                break
            else:
                print("Geçersiz seçim. Lütfen 11-16 arasında bir sayı girin.")
        except ValueError:
            print("Geçersiz seçim. Lütfen 11-16 arasında bir sayı girin.")

    # Seçilen pizza ve sos için nesneleri oluştur
    
    #Pizza
    if pizza_choice == 1:
        pizza = KlasikPizza()
    elif pizza_choice == 2:
        pizza = MargaritaPizza()
    elif pizza_choice == 3:
        pizza = TurkPizza()
    else:
        pizza = SadePizza()
    
    #Sos
    if sos_choice == 11:
        sos = Zeytin(pizza)
    elif sos_choice == 12:
        sos = Mantarlar(pizza)
    elif sos_choice == 13:
        sos = KeciPeyniri(pizza)
    elif sos_choice == 14:
        sos = Et(pizza)
    elif sos_choice == 15:
        sos = Sogan(pizza)
    else:
        sos = Misir(pizza)
    
    # Siparişin toplam fiyatını hesapla ve ekrana yazdır
    total_cost = sos.get_cost()
    tanim = sos.get_description()
    print(f"Siparişinizin toplam fiyatı: {total_cost:.2f} TL")
    print(f"Sipariş : {tanim}")


    # Kullanıcı bilgilerini al
    
    #Ad String türünde ve yalnızca harflerden oluşacak 
    while True:
     try:
        ad = input("Adınız: ")
        if not ad.isalpha():
            print("Hatalı giriş! Adınız yalnızca harflerden oluşmalıdır.")
        else:
            break
     except ValueError:
        print("Hatalı giriş! Lütfen geçerli bir değer girin.")
    
    #TCKN 16 basamaklı ve int veri türünde olacak
    while True:
     try:
        tc = int(input("TC Kimlik Numaranız: "))
        if len(str(tc)) != 11:
            print("Hatalı giriş! TC Kimlik numarası 11 haneli olmalıdır.")
        else:
            break
     except ValueError:
        print("Hatalı giriş! TC Kimlik numarası yalnızca rakamlardan oluşmalıdır.")
    
    #Kredi Kartı No 16 haneli olacak ve int karakterlerden oluşacak
    while True:
      try:
        kk_num = int(input("Kredi Kartı Numaranız: "))
        if len(str(kk_num)) != 16:
            print("Hatalı giriş! Kredi kartı numarası 16 haneli olmalıdır.")
        else:
            break
      except ValueError:
        print("Hatalı giriş! Kredi kartı numarası yalnızca rakamlardan oluşmalıdır.")
    
    #Şifre 4 haneli ve int karakterlerden oluşacak
    while True:
       try:
        kk_pswd = int(input("Kredi Kartı Şifreniz: "))
        if len(str(kk_pswd)) != 4:
            print("Hatalı giriş! Kredi kartı şifresi 4 haneli olmalıdır.")
        else:
            break
       except ValueError:
        print("Hatalı giriş! Kredi kartı şifresi yalnızca rakamlardan oluşmalıdır.")

    tarih = datetime.datetime.now()


    # Sipariş bilgilerini Orders_Database.csv dosyasına kaydet
    with open('Orders_Database.csv', mode='a', newline='\n') as csv_file:
     baslik = ['Ad', 'TC', 'Kredi_Kart_No', 'Sifre', 'Siparis_Tanım', 'Siparis_Toplam', 'Tarih']
     writer = csv.DictWriter(csv_file, fieldnames=baslik)
     csv_file.seek(0)
     ##writer.writeheader() ilk çalıştırmada kullanıldı
     writer.writerow({'Ad': ad, 'TC': tc, 'Kredi_Kart_No': kk_num, 'Sifre': kk_pswd, 'Siparis_Toplam': total_cost, 'Siparis_Tanım': tanim, 'Tarih': tarih})
```

```python
main()
```
